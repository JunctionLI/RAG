```python
from typing import List

def split_into_chunks(doc_file: str) -> List[str]:
    with open(doc_file, 'r', encoding='utf-8') as file:  ###打开名为 `doc_file` 的文件，使用只读模式（`'r'`）；指定编码为 `'utf-8'`，防止乱码或编码错误  `with ... as ...` 是一个上下文管理器，会在读取完毕后自动关闭文件
        content = file.read()

    return [chunk for chunk in content.split("\n\n")]

chunks = split_into_chunks("doc.md")

for i, chunk in enumerate(chunks):  #使用 `for` 循环遍历 `chunks` 列表；`enumerate()` 同时返回索引（`i`）和内容（`chunk`）
    print(f"[{i}] {chunk}\n")
```

```python
from sentence_transformers import SentenceTransformer ##从 sentence-transformers 库中导入用于加载文本嵌入模型的类。

embedding_model = SentenceTransformer("all-MiniLM-L6-v2") ##这行代码会联网去 Hugging Face 下载一个能将英文句子转为向量的AI模型，供你后续使用。
def embed_chunk(chunk: str) -> List[float]:
    embedding = embedding_model.encode(chunk)
    return embedding.tolist()

test_embedding = embed_chunk("test content")
print(len(test_embedding))
print(test_embedding)
```

```python
embeddings = [embed_chunk(chunk) for chunk in chunks]
print(len(embeddings))
print(embeddings[0])
```

```python
import chromadb  ##向量数据库

chromadb_client = chromadb.EphemeralClient()   ##创建客户端，内存型数据库，不会存入磁盘，脚本运行结束就自动清除数据  想要存磁盘，用 chromadb.PersistentClient("./chroma.db")  会存在chroma.db这个文件里面
chromadb_collection = chromadb_client.get_or_create_collection(name="default")  #可以理解成传统数据库的表格

def save_embeddings(chunks: List[str], embeddings: List[List[float]]) -> None:
    ids = [str(i) for i in range (len(chunks))]   #为每个 chunk 创建一个唯一的字符串 ID，Chroma 要求所有数据项都有唯一 ID（类似数据库的主键 primary key）
    chromadb_collection.add(
        documents=chunks,
        embeddings=embeddings,
        ids=ids
    )
#对应内容全部存入向量数据库


save_embeddings(chunks,embeddings)

```

```python
def retrieve(query:str, top_k: int) -> List[str]:  #top_k指的是返回前几个最相关的结果
    query_embedding = embed_chunk (query) #把用户的问题转换为向量，让计算机对比语义
    results = chromadb_collection.query(
        query_embeddings = [query_embedding],  #查询时必须是一个列表
        n_results = top_k  #n_results 是 Chroma 向量数据库 的查询函数 query() 所必需的参数之一。每个查询问题，你最多要返回几个最相似的文本结果？
    )
    return results['documents'][0]

query = "who save the villagers?"
retrieved_chunks = retrieve(query,5)

for i, chunk in enumerate(retrieved_chunks):
    print(f"[{i}]{chunk}\n")
```

```python
from sentence_transformers import CrossEncoder

def rerank(query:str, retrieved_chunks: List [str], top_k: int) -> List [str]:
    cross_encoder = CrossEncoder('cross-encoder/mmarco-mMiniLMv2-L12-H384-v1')
    pairs = [(query, chunk) for chunk in retrieved_chunks]  #用户问题和召回片段组成pair
    scores = cross_encoder.predict(pairs)  #发给encoder给用户问题和召回片段相似度打分

    chunk_with_score_list = [(chunk,score)
                         for chunk, score in zip(retrieved_chunks, scores)]
    chunk_with_score_list.sort(key=lambda pair: pair[1], reverse=True)  #从高到低（降序）排序，分数越高 → 越靠前，越相关的段落 → 越先被返回； Lamda 指的是匿名函数，临时写个小函数，只有一行代码，不用取名字
                                                                        #lamda --- def get_score(pair):return pair[1]
    return [ chunk for chunk, _ in chunk_with_score_list][:top_k]

reranked_chunks = rerank(query, retrieved_chunks, 3)

for i, chunk in enumerate(reranked_chunks):
    print(f"[{i}] {chunk}\n")
```

```python
from dotenv import load_dotenv # 从 python-dotenv 包中导入 load_dotenv 函数。
from google import genai #导入 Google 的 genai 模块，这个模块是用来访问 Gemini API 的。

load_dotenv()  ##把env变成环境变量，Gemini会从环境变量里面找到API KEY
google_client= genai.Client() #创建一个 Gemini 的客户端对象（用于调用 Gemini 模型）

def generate(query:str, chunks:List[str]) -> str: #创建一个 Gemini 的客户端对象（用于调用 Gemini 模型）
    prompt = f"""You are knowledge assistant, please give an accurate answer based on user's question and the following chunk,     

user's quesion:{query}
the chunk:
{"\n\n".join(chunks)}

please based on above content to answer, don't make up the fake information."""    # 把 chunks 列表里的多段文字，用两个换行符 \n\n 连接起来，合成一段新的文字。
                                                                                    #这样能让模型更清楚每一段内容的结构，让它更容易理解文本背景和分段信息。

    print(f"{prompt}\n\n---\n")

    response = google_client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return response.text

answer = generate(query, reranked_chunks)
print(answer)
```
